### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Pyplot style or the object oriented way. The recommended way is the object oriented way. Since it gives more control and flexability while the pyplot way might be faster.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: Is the parent which includes the axes, axis and artists. It holds everything that u see in a plot.
Axes: Is the container for one graph. The figure can contain several Axes.
Axis: Axises are the x/y in a 2d plot or the x/y/z in a 3D plot.
Artist: Artists are everything that is displaying. Text, lines, shapes, axes
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Sequence of numbers
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

plt.plot(x, y)
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
