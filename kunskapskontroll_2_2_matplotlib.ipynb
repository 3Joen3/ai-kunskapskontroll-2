```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Pyplot style or the object oriented way. The recommended way is the object oriented way. Since it gives more control and flexability while the pyplot way might be faster for simple plots.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: Is the parent which holds everything in the plot. Including axes, axis and artists.
Axes: Is the container for one graph, one coordinate system. The figure can contain multiple Axes.
Axis: X and Y in a 2d plot or X, Y and Z in a 3D plot.
Artist: Artists are everything that is drawn. Text, lines, shapes and axes.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
A sequence of numbers
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

plt.plot(x, y)
plt.title("y = x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))

ax[0].set_title("Scatter plot")
ax[0].scatter(random_data_x,random_data_y, color="yellow")

ax[1].set_title("Bar plot")
ax[1].bar(names, values, color="purple")
ax[1].set_xlabel('Fruit')
ax[1].set_ylabel('Count')

plt.show()
```
